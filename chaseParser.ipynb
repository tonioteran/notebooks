```python
import pandas as pd
from tabula import read_pdf
import subprocess
from subprocess import Popen, PIPE
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
def chase2df(pdf_file,year):
    command  = ['pdftotext', '-layout', pdf_file, '-']
    output   = subprocess.check_output(command).decode()
    colnames = ['Year', 'Date', 'Description', 'Amount']
    expenses = pd.DataFrame(columns=colnames)
    payments = pd.DataFrame(columns=colnames)
    for item in re.findall(r'(?m)^[ ]*[0-9]{2}\/[0-9]{2}.*\.[0-9]{2}$', output):
        data = item.split()
        tmpdf = pd.DataFrame([[int(year), data[0], " ".join(data[1:-1]), float(data[-1].replace(',',''))]], columns = colnames)
        if tmpdf.loc[0,'Amount'] < 0:
            payments = payments.append(tmpdf, ignore_index=True)
        else:
            expenses = expenses.append(tmpdf, ignore_index=True)
    return expenses,payments
    
```

```python
# set path to directories
master_dir = "/home/tonio/Documents/bankStatements/sapphire/"
# arrays for saving parsed dfs
all_expenses = []
all_payments = []
```

```python
colnames = ['Year', 'Date', 'Description', 'Amount']
expenses_master = pd.DataFrame(columns=colnames)        
payments_master = pd.DataFrame(columns=colnames)        
for year in subprocess.check_output(["ls",master_dir]).decode().split():
    for statement in subprocess.check_output(["ls", master_dir+year+"/"]).decode().split():
        print(statement)
        tmp_exp, tmp_pay = chase2df(master_dir+year+"/"+statement,year)
        expenses_master = expenses_master.append(tmp_exp, ignore_index=True)
        all_expenses.append(tmp_exp)
        payments_master = payments_master.append(tmp_pay, ignore_index=True)
        all_payments.append(tmp_pay)
```
